```python
new_movies['genres'] = new_movies['genres'].apply(lambda X:X.lower())
new_movies.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_movies['genres']=new_movies['genres'].apply(stem)
```

# Collaborative Filtering Based Recommender System

```python
ratings_data.head()
```

```python
final_dataset = ratings_data.pivot(index='movieId',columns='userId',values='rating')
final_dataset.head()
```

```python
final_dataset.fillna(0,inplace=True)
final_dataset.head()
```

```python
from scipy.sparse import csr_matrix

csr_data = csr_matrix(final_dataset.values)
final_dataset.reset_index(inplace=True)
```

```python
final_dataset.head()
```

```python
user = int(input('Enter the userID: '))
N = int(input('Enter the number of recommendations you want: '))
k = int(input('Enter the threshold for number of users: '))
```

```python

from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20, n_jobs=-1)
knn.fit(csr_data)
```

```python
def get_movie_recommendation(user, N, k):
    n_movies_to_recommend = N
    movie_list = ratings_data[ratings_data['userId'] == user]
    if not movie_list.empty:
        movie_idx = movie_list.iloc[0]['movieId']
        movie_idx = final_dataset[final_dataset['movieId'] == movie_idx].index[0]
        distances, indices = knn.kneighbors(csr_data[movie_idx], n_neighbors=n_movies_to_recommend + 1)
        rec_movie_indices = sorted(list(zip(indices.squeeze().tolist(), distances.squeeze().tolist())), key=lambda x: x[1])[:0:-1]
        recommend_frame = []
        for val in rec_movie_indices:
            movie_idx = final_dataset.iloc[val[0]]['movieId']
            idx = movies_data[movies_data['movieId'] == movie_idx].index
            recommend_frame.append({'Title': movies_data.at[idx[0], 'title'], 'Distance': val[1]})  # Use .at to access the title
        df = pd.DataFrame(recommend_frame, index=range(1, n_movies_to_recommend + 1))
        return df
    else:
        return "No movies found. Please check your input"

recommendations = get_movie_recommendation(user, N, k)
print(recommendations)
```
